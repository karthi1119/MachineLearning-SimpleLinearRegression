```python
#To use a python library(pandas) to use functions that are inbuilt in pandas
import pandas as pd  #pandas renamed as pd
```

```python
#variable is assigned to pd and to read the csv file(sample csv file)
dataset=pd.read_csv("sample.csv")
```

```python
dataset #print
```

```python
#to split the trainset and testset
#variable is independent
independent=dataset[["YearExperience"]]
independent
```

```python
#variable is dependent assigned to salary
dependent=dataset[["Salary"]]
dependent
```

```python
#here python library is used (.)is a another function to call 
# train test split is class name 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0)  #(x_train,x_text,y_train,y_test are variables to arrange the values)
```

```python
#linear model is another function name and to import the LinearRegression function
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()   #regressor is variable to store the functions into this(from linearRegression())
regressor.fit(x_train,y_train)  #to fit the(x_train,y_train) inputs in linearRegression
```

```python
#to calculate weight from regressor (coef_)
weight=regressor.coef_
weight
```

```python
#to calculate bais from regressor (intercept_)
bais=regressor.intercept_
bais
```

```python
#y_pred is variable
y_pred=regressor.predict(x_test)  #x_test is the input of predicted variable
```

```python
# metrics is python library function and r2_score is also a function of python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)    #r_score is variable,(y_test,y_pred) are actual and predicted values inserted to calculate r_score
r_score  # the r_score is nearly to 1 is the best AI program else nearly to 0 is poor AI program
```

```python
# to save the module we use pickle function
import pickle
```

```python
pickle.dump(regressor,open(filename,'wb'))   #dump the full function into pickle variable   
                                             # ('wb') is write binary 
```

```python
# variable created and it is the end user applicattion to use frequent 
# 'rb' is read binary to read and to get the input
loaded_model=pickle.load(open("finalized_model_linear.sav",'rb'))
```

```python
#vaiable is assigned to program and it was the final input to get the output of the AI program 
result=loaded_model.predict([[13]])       #load_model has full function to calculate the given integer and it will predict
```
